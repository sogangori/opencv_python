## 이미지 피라미드

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
%matplotlib inline
```

```python
import matplotlib as mpl
mpl.rcParams['image.cmap'] = 'gray'
```

```python
!dir
```

```python
img = cv2.imread('frog.jpeg')
img.shape
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray)
```

```python
down = cv2.pyrDown(gray)
print(down.shape)
down = cv2.pyrDown(down)
print(down.shape)
down = cv2.pyrDown(down)
print(down.shape)
```

```python
plt.imshow(down)
```

```python
up = cv2.pyrUp(gray)
print(up.shape)
up = cv2.pyrUp(up)
print(up.shape)
up = cv2.pyrUp(up)
print(up.shape)
```
