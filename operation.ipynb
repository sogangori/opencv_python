bit 연산자

```python
img2gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img2gray.shape
```

```python
plt.imshow(img2gray, cmap='gray')
plt.colorbar()
```

```python
a = np.arange(256, dtype=np.uint8).reshape((16,16))
a.shape
```

```python
ret, mask = cv2.threshold(img2gray, 100, 255, cv2.THRESH_BINARY)
ret, mask.shape, np.unique(mask)
```

```python
plt.imshow(mask, cmap='gray')
plt.colorbar()
```

```python
plt.imshow(cv2.bitwise_not(mask), cmap='gray')
plt.colorbar()
```

```python
img1_bg = cv2.bitwise_and(img2gray, img2gray, mask=mask)
img1_bg.shape
```

```python
plt.imshow(img1_bg, cmap='gray')
```

성능 측정

```python
e1 = cv2.getTickCount()
k = cv2.countNonZero(img2gray)
e2 = cv2.getTickCount()
time = (e2 - e1)/cv2.getTickFrequency()
print('time', time)
```
