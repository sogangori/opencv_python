```python
import pytesseract
```

```python
import matplotlib as mpl
mpl.rcParams['image.cmap'] = 'gray'
```

```python
img = cv2.imread('korea.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(gray)
```

```python
string = pytesseract.image_to_string(gray, lang='kor')
print(string)
```
