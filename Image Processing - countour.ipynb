## 윤곽선 contour

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
%matplotlib inline
```

```python
import matplotlib as mpl
mpl.rcParams['image.cmap'] = 'gray'
```

```python
!dir
```

```python
#img = cv2.imread('A.png')
img = cv2.imread('apple.jpg')
img.shape
img = cv2.resize(img, (300,300))
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray)
```

```python
ret, thresh = cv2.threshold(gray, 230, 255, 0)
```

```python
plt.imshow(thresh)
```

```python
contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
len(contours), len(hierarchy)
```

```python
con = cv2.drawContours(img, contours, -1, (255,0,0), 3)
plt.imshow(con)
```

```python
cnt = contours[1]
cnt.shape
```

```python
canvas = np.zeros_like(img)
for i in range(len(cnt)):
    point = tuple(cnt[i][0])    
    canvas = cv2.circle(canvas,point,5,[255,0,0],-1)
```

```python
plt.imshow(canvas)
```

```python
cnt = contours[1]
M = cv2.moments(cnt)
```

```python
area = cv2.contourArea(cnt)
perimiter = cv2.arcLength(cnt, closed=True)
'면적', area,'둘레 길이', perimiter
```
